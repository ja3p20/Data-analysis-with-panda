```python
bond = pd.read_csv("pandas/jamesbond.csv", index_col = "Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Bresnan":
        return "The best"
    elif (actor == "Roger Moore" and budget > 40):
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie,axis="columns")  #note that we use or traversing columns based perspective
```
