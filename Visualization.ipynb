```python
goog = data.DataReader(name="GOOG", data_source="yahoo", start= "2004-01-01", end= "2016-12-31")
goog.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price < 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
goog["Close"].apply(rank_performance).value_counts()
```

```python
goog["Close"].apply(rank_performance).value_counts().plot(kind = "barh") # barh for horizontal bar
```

```python
goog["Close"].apply(rank_performance).value_counts().plot(kind = "pie")
```

```python
apple = data.DataReader(name="AAPL", data_source="yahoo", start="2012-01-01", end= "2016-12-31")
apple.head()
```

```python
apple["Close"].mean()
```

```python
def rank_perform(stockprice):
    if stockprice >= 93.3617192254726:
        return "Above Average"
    else:
        return "Below Average"
```

```python
apple["Close"].apply(rank_perform).value_counts().plot(kind = "pie", legend = True)
```
